## Question 1

Define a procedure called "check_input" that takes an input as a square list of lists (2D Matrix of N x N) and returns a boolen "true" if the input is valid or "false" otherwise, according to the following criteria.

1. Each coloumn must contain numbers from between 1 and N, with no duplicates.
2. Each row must contain numbers from between 1 and N, with no duplicates.

```python
def check_input(array):
    r = 0
    c = 0
    x = len(array[0])
    for i in range(0,x):
        for j in range(1,x):
            if(array[i][0]==array[i][j] or type(array[i][j])==str):
                r = r + 1
    
    for j in range(0,3):
        for i in range(1,3):
            if(array[0][j]==array[i][j] or type(array[i][j])==str):
                c = c+ 1
    if(r==0 and c==0):
        return True
    else:
        return False
```

```python
# run this block to test code after defining the procedure

test1 = [[1, 2, 3],
        [2, 3, 1],
        [3, 1, 2]]

test2 = [[1, 2, 3, 4],
        [2, 3, 1, 3],
        [3, 1, 2, 3],
        [4, 4, 4, 4]]

test3 = [['a', 'b', 'c'],
        ['b', 'c', 'a'],
        ['c', 'a', 'b']]

print(check_input(test1)) # must return true
print(check_input(test2)) # must return false
print(check_input(test3)) # must return false
```

**Part A** - From the DataFrame created in Question 2, calculate the average marks for each subject

```python
import pandas as pd
import numpy as np

df= pd.read_csv("StudentMarkSheet.csv")

df=df.dropna(how='any')

df[ 'Exam name'].unique()

p=df[df[ 'Exam name']=='Philosophy']['Marks'].mean()

print('Philosophy: ', p)
ch=df[df[ 'Exam name'] == 'Chemistry']['Marks'].mean()

print("Chemistry: ",ch) 
bi=df[df['Exam name']=='Biology']['Marks'].mean() 
print("Biology: ",bi) 
m=df[df[ 'Exam name']=='Mathematics']['Marks'].mean() 
print("Mathematic: ",m) 
ph=df[df[ 'Exam name']=='Physics']['Marks'].mean() 
print("Physics: ",ph) 
so=df[df[ 'Exam name'] == 'Sociology']['Marks'].mean()
print("Sociology: ", so)
```

**Part B** - Using the averages you just calculated, plot a Bar Graph showing the average marks of each subject

```python
df=pd.DataFrame({'x':['Philosophy', 'Chemistry', 'Biology', 'Mathematics', 'Physics','Sociology'],'y':[p,ch,bi,m,ph,so]})
ax=df.plot.bar(x='x',y='y',rot=0)

```

**Part C** - Create a linear regression model using sklearn and fit it to the data you just imported


```python
reg = linear_model.LinearRegression()
train_x = np.array(train[["Tail Length (cm)"]])
train_y = np.array(train[["Mass (kg)"]])
reg.fit(train_x,train_y)
```
